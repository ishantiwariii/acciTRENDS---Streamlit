## Here is some analysis on acciTRENDS dataset

```python
import numpy as np 
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'RS_Session_265_AU_1075_A_i_0.csv')
df
```

```python
df.columns
df.drop('Sl. No.', axis=1, inplace=True)
df.set_index('States/UTs', inplace=True)
```

```python
df
```

```python
df.describe()
```

```python
df.info()
```

```python
df.fillna(0, inplace=True)
df.info()
```

### changing data type of years to integer 

```python
df[['2018' , '2019' , '2020', '2021', '2022']] = df[['2018' , '2019' , '2020', '2021', '2022']].astype(int)
```

```python
df.info()
```

```python
df = df[df.index != 'Total (All India)']
```

### splitting the dataset for saperating union territories and states

```python
uts = [
    'Andaman and Nicobar Islands',
    'Chandigarh',
    'Dadra and Nagar Haveli',
    'Daman and Diu',
    'Delhi',
    'Jammu and Kashmir',
    'Ladakh',
    'Lakshadweep',
    'Puducherry'
]
```

```python
df_ut = df[df.index.isin(uts)].copy()
df_ut.index.name = 'UTs'
df_ut
```

```python
df_states = df[~df.index.isin(uts)].copy()
df_states.index.name = 'State'
df_states
```

### Now we have saperate datasets for both states and union territories acciedent 
- df_ut  # union territory dataset 
- df_states # states dataset

## ` now we will perform EDA on states dataset `

- find total deaths in 5 yrs 
- line graph using plotly for deaths across yrrs in every state
- top 10 dangerous states in terms of accident 
- top 10 safest states in term of accident 
- growth in accident for each states 
- bar graph for all states total deaths count 
- heatmap across every states for every year 
- comparing states by using line graph

```python
df_states.info()
```

```python
df_states.describe()
```

```python
df_states.head()
```

```python
df_states['Total_5yr'] = df_states.sum(axis=1)
df_states.sort_values('Total_5yr', ascending=False).head(10)
```

## line graph to compare accidents accross years 

```python

px.line(df_states.reset_index(), x='State', y=['2018','2019','2020','2021','2022'])
```

## top 10 states in case of accidents

```python
top = df_states['Total_5yr'].sort_values(ascending=False).head(10)
px.bar(top.reset_index(), x='State', y='Total_5yr', color='State')
```

## safest 10 states

```python
least_10 = df_states['Total_5yr'].sort_values().head(10)
px.bar(least_10.reset_index(), x='State', y='Total_5yr', color='State', log_y=True)
```

## Growth in accidents through all years 

```python
df_states['Growth_%'] = ((df_states['2022'] - df_states['2018'])/df_states['2018'])*100
df_states
```

## bar graph for all states analysis

```python
df_states_plot = df_states.reset_index()
df_states_plot.rename(columns={'index': 'State'}, inplace=True)
px.bar(df_states_plot, x='State' , y='Total_5yr' , color='State',log_y=True)
```

```python
sns.heatmap(df_states, cmap='Reds')
```

## comparing the states and for that we use loc for extracting the rows and then T for transpose of rows and columns

```python
df_states.loc[['Tamil Nadu','Uttar Pradesh']].T.plot()
```

```python
df_states.loc[['Mizoram','Sikkim']].T.plot()
```
